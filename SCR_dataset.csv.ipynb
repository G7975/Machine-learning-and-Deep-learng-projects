```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostRegressor
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("scr-dataset.csv")
```

```python
df.shape
```

```python
df.plot(x = 'x', y = 'y')
plt.show()
```

```python
df.corr()
```

```python
x, y = df['x'],df['y']
```

```python
x,y = pd.DataFrame(x), pd.DataFrame(y)
```

```python
model = DecisionTreeRegressor(max_depth =20, criterion="squared_error")
model.fit(x.values, y)
```

```python
model.score(x.values,y)
```

```python
model_pred = model.predict(x.values)
```

```python
model_pred.shape
```

```python
plt.figure()
plt.scatter(x,y, label = 'Training data "x"')
plt.plot(x, model_pred,c='g', label = 'DecisionTreeRegressor_pred',linewidth = 2)
plt.xlabel('X_train')
plt.ylabel('Y_target')
plt.show()
```

```python
model.predict([[50]])
```
